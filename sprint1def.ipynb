Full integration of the tables: trip information and rent and population information about the destiny and the origin

```python
con.execute('''
CREATE OR REPLACE TABLE silver_integrated_od AS
SELECT 
    t.date,
    t.hour,
    t.n_trips,
    t.trips_total_length_km,
    dayname(t.date) AS day_of_week,
    (dayofweek(t.date) IN (0, 6)) AS is_weekend,
    t.id_origin AS origin_code_mitma,
    o.zone_name AS origin_name,
    o.population AS origin_population,
    o.rent AS origin_rent,
    t.id_destination AS destination_code_mitma,
    d.zone_name AS destination_name,
    d.population AS destination_population,
    d.rent AS destination_rent

FROM all_trips t
JOIN bronze_rel_towns rel_o ON t.id_origin = rel_o.code_mitma
JOIN bronze_rel_towns rel_d ON t.id_destination = rel_d.code_mitma
JOIN silver_zone_metrics o ON rel_o.code_ine = o.zone_id
JOIN silver_zone_metrics d ON rel_d.code_ine = d.zone_id
WHERE t.n_trips >= 0;
''')

con.execute("SELECT * FROM silver_integrated_od LIMIT 5").df()
```

GOLDEN PHASE

Ready-to-use model

```python
con.execute('''
CREATE OR REPLACE TABLE gold_analysis AS
WITH calc AS (
    SELECT 
        v.origin_code_mitma AS origin_id,
        v.n_trips AS total_trips,
        p1.zone_name AS origin_name,
        (CAST(p1.population AS BIGINT) * CAST(p2.population AS BIGINT)) /
        NULLIF(POWER(v.trips_total_length_km / NULLIF(v.n_trips, 0), 2), 0) AS theoretic_demand
    FROM silver_integrated_od v
    JOIN silver_zone_metrics p1 ON v.origin_code_mitma = p1.zone_id
    JOIN silver_zone_metrics p2 ON v.destination_code_mitma = p2.zone_id
    WHERE v.origin_code_mitma != v.destination_code_mitma
      AND v.n_trips > 0
)

SELECT 
    origin_id,
    origin_name,
    SUM(total_trips) AS total_trips,
    CAST(SUM(theoretic_demand) AS BIGINT) AS theoretic_demand,
    ROUND((SUM(total_trips) / NULLIF(SUM(theoretic_demand), 0)) * 1000000, 2) AS cover
FROM calc
GROUP BY origin_id,origin_name
HAVING total_trips > 500
ORDER BY cover ASC;
''')

con.execute("""
    SELECT *
    FROM gold_analysis
    LIMIT 10
""").fetchdf()
```

¿Qué zonas con una población mayor a 20k tienen peor servicio de infraestructura?

```python
con.execute("""
SELECT
    g.origin_id,
    g.origin_name,
    p.population,
    g.total_trips,
    g.cover
FROM gold_analysis g JOIN silver_zone_metrics p ON g.origin_id = p.zone_id
WHERE p.population > 20000
ORDER BY g.cover ASC
LIMIT 10;
""").df()
```

```python
con.execute('''
CREATE OR REPLACE TABLE gold_flujos_provinciales AS
SELECT 
    po.name AS provincia_origen,
    pd.name AS provincia_destino,
    SUM(v.n_trips) AS total_trips,
    ROUND(SUM(v.trips_total_length_km) / NULLIF(SUM(v.n_trips), 0), 2) AS distancia_media_km
FROM silver_integrated_od v
JOIN provinces po  ON SUBSTR(v.origin_code_mitma, 1, 2) = po.code
JOIN provinces pd  ON SUBSTR(v.destination_code_mitma, 1, 2) = pd.code
WHERE po.code != pd.code
GROUP BY provincia_origen, provincia_destino
ORDER BY total_trips DESC;
''')

con.execute("SELECT * FROM gold_flujos_provinciales LIMIT 5").df()
```

Gráfico viendo patrones de tráfico en Valencia:

```python
import pandas as pd
import duckdb
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

df_valencia_hourly = con.execute("""
SELECT
    v.hour,
    SUM(v.n_trips) AS total_trips
FROM silver_integrated_od v
JOIN provinces p
    ON SUBSTR(v.origin_code_mitma, 1, 2) = p.code
WHERE p.name = 'Valencia'
GROUP BY v.hour
ORDER BY v.hour
""").df()


plt.figure(figsize=(12, 7))
def format_y_axis(x, pos):
    if x >= 1e6:
        return f'{x*1e-6:1.1f}M'
    elif x >= 1e3:
        return f'{x*1e-3:1.0f}K'
    return f'{x:1.0f}'

sns.scatterplot(
    x='hour',
    y='total_trips',
    data=df_valencia_hourly,
    s=150, 
    label='Total trips'
)

sns.lineplot(
    x='hour',
    y='total_trips',
    data=df_valencia_hourly,
    alpha=0.4,
    linewidth=2)

plt.title('Movilidad Total por Hora en la Provincia de Valencia', fontsize=16)
plt.xlabel('Hora del Día (00:00 - 23:00)', fontsize=14)
plt.ylabel('Total de Viajes Enviados', fontsize=14)
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_y_axis))
plt.xticks(range(0, 24), rotation=0)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.legend(loc='upper left')
```

Ahora faltaría añadir los datos del INE

Probamos a crear una tabla de prueba para ver los datos:

```python
import duckdb

con = duckdb.connect()

# Registra el DataFrame como una vista temporal en DuckDB
con.register('df_bronze', df_bronze)

# Usa SQL para crear o reemplazar la tabla a partir de la vista temporal
con.execute("""
    CREATE OR REPLACE TABLE bronze_mitma_prueba AS SELECT * FROM df_bronze
""")

con.execute("""
    SELECT * FROM bronze_mitma_prueba LIMIT 50
""").fetchdf()
```

Probamos a mostrar 3 trayectos por zona ordenado de mayor a menor distancia recorrida para ver que todo haya funcionado:

```python
con.execute("""
    SELECT *
    FROM (
        SELECT *,
            ROW_NUMBER() OVER (
                PARTITION BY zone_level ORDER BY trips_total_length_km DESC
            ) AS rn
        FROM bronze_mitma_prueba
    )
    WHERE rn <= 3
    ORDER BY zone_level, trips_total_length_km DESC
""").fetchdf()
```

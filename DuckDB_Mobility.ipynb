```python
import duckdb
con = duckdb.connect('./mobility.db')
con.execute(""" ---sql
CREATE OR REPLACE TABLE trips (
    date VARCHAR,
    hour BIGINT,
    id_origin VARCHAR,
    id_destination VARCHAR,
    n_trips DOUBLE,
    trips_total_length_km DOUBLE
);
""")
con.execute(""" ---sql
INSERT INTO trips (date, hour, id_origin, id_destination, n_trips, trips_total_length_km)
SELECT date, hour, id_origin, id_destination, n_trips, trips_total_length_km
FROM read_parquet('C:/Users/carlo/data/Viajes_GAU_2024-10-14_2024-10-20_v2.parquet');
""").fetchdf()
```

```python
# Listar todas las filas (¡solo primeras 10 para no saturar!)
con.execute(""" ---sql
    SELECT * 
    FROM trips 
    ORDER BY trips_total_length_km ASC
    LIMIT 10
""").fetchdf()
```
